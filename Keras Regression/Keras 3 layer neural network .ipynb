## Neural Network Using Keras !!
We will create a 3 layer neural network using Keras .

```python
# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import sys
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

```python
# load pima indians dataset
dataframe = pd.read_csv("pima-indians-diabetes.csv")
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

<span style="color: #181407 ; font-family: Babas;">Models in Keras are defined as a sequence of layers. We create a Sequential model and add layers one at a time until we are happy with our network topology. The first thing to get right is to ensure the input layer has the right number of inputs. This can be specified when creating the first layer with the input dim argument and setting it to 8 for the 8 input variables</span> 

```python
# create model
model = Sequential()
 
# In this example we will use a fully-connected network structure with three layers.

model.add(Dense(12, input_dim=8, kernel_initializer= 'uniform' , activation = 'relu' ))
model.add(Dense(8, kernel_initializer= 'uniform' , activation= 'relu' ))
model.add(Dense(1, kernel_initializer= 'uniform' , activation= 'sigmoid' ))
```

<span style="color: #181407; font-family: Babas;">We can piece it all together by adding each layer. The first hidden layer has 12 neurons and expects 8 input variables. The second hidden layer has 8 neurons and finally the output layer has 1 neuron to predict the class as it is sigmoid function. It will predict the probablity from 0 to 1</span>

###  Compiling the model !!

Compiling the model uses the efficient numerical libraries under the covers (the so-called backend) such as Theano or TensorFlow.

```python
# Compile model
model.compile(loss= 'binary_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ])
# Fit the model
model.fit(X, Y,validation_split=0.33, epochs=300, batch_size=10)
# evaluate the model
scores = model.evaluate(X, Y)
if 
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
